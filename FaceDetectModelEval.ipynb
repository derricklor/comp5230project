<h1> Code to evaluate the face detect architecture

```python
import os
import numpy as np
import hog as hg
import faceAlign as face
import randomSelectImages as rsi

def reload_known_faces_list(path):
    hog_list = []
    name_list = []
    # check if path exists, if not then make path
    if not os.path.exists(path):
        print(f"Path {path} does not exist, creating path now")
        os.makedirs(path)

    # create encoding and name from files in the path
    for filename in os.listdir(path):
        if filename.endswith(('.jpg', '.png')):
            image_path = os.path.join(path, filename)

            # align face from path
            try:
                face_frame = face.align_face_from_file(image_path)
            except Exception as e:
                print(f'Error with {filename}')

            # calculate hog of aligned face frame
            image_hog, _ = hg.calculate_hog_from_face_frame(face_frame) # dont need visualizations

            hog_list.append(image_hog)
            name = os.path.splitext(filename)[0]
            name_list.append(name)
            
    return hog_list, name_list


celeb_faces_path = r"C:\Users\Derrick\Documents\School\Computer Vision\project\known_faces\celeb faces"

# initialize face library, apppend "face" class folder
whitelist_hogs, whitelist_names = reload_known_faces_list(celeb_faces_path + r"\face")

```

```python
def runBatch(num_images, method, threshold, output=False):
    # randomly select celeb face images from dataset
    # "C:\Users\Derrick\Documents\School\Computer Vision\project\known_faces\celeb faces\face"
    celeb_paths = rsi.select_random_images(r"C:\Users\Derrick\Documents\School\Computer Vision\project\datasets\Celeb Images", 10)

    # init some global variables
    distance_method = method # 'euclidean', 'cosine', or 'correlation'
    distance_threshold = threshold #0.17
    correct = 0

    celeb_hog_list = []
    celeb_name_list = []

    # Loop through each celeb faces in celeb_paths
    for celeb_path in celeb_paths:
        # try align face from single celeb
        try:
            face_frame = face.align_face_from_file(celeb_path)
        except Exception as e:
            print(f'Error with {celeb_path}')

        if face_frame is not None:
            # calculate hog of aligned face frame
            celeb_hog, _ = hg.calculate_hog_from_face_frame(face_frame) # dont need visualizations

            celeb_hog_list.append(celeb_hog)
            celeb_name = os.path.basename(celeb_path)
            celeb_name = celeb_name.split(" ")[0] # get the first name of celeb
            celeb_name_list.append(celeb_name)
        else:
            print(f"Failed to detect and recognize face from {celeb_path}.")

    for i in range (len(celeb_hog_list)):
        total = len(celeb_hog_list)
        # reset variables
        whitelist_dist = []
        name = "unknown"
            
        # if face imgs exist in directory then do comparison of each face with celeb face
        if whitelist_hogs is not None and whitelist_hogs:
            for hog in whitelist_hogs:
                
                whitelist_dist.append(hg.compare_hog_features(hog, celeb_hog_list[i], distance_method))
                try:   
                    # then check whitelist for matches
                    if (len(whitelist_dist) > 0): #check if not empty
                        lowest_score = np.min(whitelist_dist)
                        # if score is below threshhold then assign whitelist face
                        if lowest_score < distance_threshold:
                            best_index = np.argmin(whitelist_dist)
                            name = whitelist_names[best_index]
                            name = name.split(" ")[0] # get the first name of celeb
                except Exception as e:
                    print(f"Error in lists: {e}")
        if output is True:
            print(f' Celeb truth name is {celeb_name_list[i]}, the predicted name is {name}.')
        if celeb_name_list[i] == name:
            correct+=1
    
    accuracy = correct/total
    print(f'Total images: {total}')
    print(f'Correct: {correct}, Accuracy: {accuracy}')

    return accuracy

                        
```

```python
total = 0
correct = 0
num_img = 10
iterations = 10
distance_method = ["euclidean", "cosine", "correlation"]
distance_threshold = [9, 0.17, 1.1]
```

```python
for k in range(3):
    acc_list = []
    for i in range(iterations):
        acc = runBatch(num_img, distance_method[k], distance_threshold[k], output=False) #default output flag is false, true for printing output
        acc_list.append(acc)
    print(f'Using distance method: {distance_method[k]} with threshold: {distance_threshold[k]}.')
    print(f'Averaged accuracy over {iterations} iterations: {np.mean(acc_list)}')
```
